**1.Currently How Many Stocks are Available?**


```python
stocks_available=df[df['In_Stock'] == 'Yes'].shape[0]
print('Total number of shoes currently in stock:',stocks_available)
```

This code filters the DataFrame df to count the rows where the 'In_Stock' column is 'Yes', indicating
shoes that are available. The .shape[0] returns the number of such rows, representing the total shoes in
stock, and the print() statement outputs this count with a message.

**2.What is the Avg Price of Reebok Shoes?**

```python
average_price_of_Reebok_Shoes = df[df['Brand'] == 'Reebok']['Price'].mean()
print('Average price of Reebok shoes:',average_price_of_Reebok_Shoes)
```

This code calculates and prints the average price of Reebok shoes. It first filters the DataFrame df to
include only rows where the 'Brand' is 'Reebok'. Then, it selects the 'Price' column of the filtered rows
and uses .mean() to compute the average price. Finally, the print() statement displays the calculated
average price along with a descriptive message.

**3.Is there Relationship in Rating & Price Justify**


```python
sns.regplot(x='Price', y='Rating', data=df)
plt.title('Scatter Plot of Price vs Rating with Regression Line')
plt.show()
```

This code creates a scatter plot with a regression line showing the relationship between the 'Price' and
'Rating' columns in the DataFrame df. The sns.regplot() function from Seaborn plots the scatter points
for each pair of 'Price' and 'Rating' values and adds a linear regression line to indicate the trend. The
plt.title() adds a title to the plot, and plt.show() displays the plot. This visual helps in understanding
how the price of an item correlates with its rating, along with the trend line for better clarity

**4.How much cost required to complete fill the pending stock**

```python
pending_stock = df[df['In_Stock']=='No']
total_cost = pending_stock['Price'].sum()
print('Total cost required to complete the pending stock',total_cost)
```

This code calculates and prints the total cost required to complete the pending stock of items in the
DataFrame df. It first filters the DataFrame to include only rows where the 'In_Stock' column is 'No',
representing items that are out of stock. Then, it sums up the 'Price' values of these pending stock
items using .sum(). The print() statement displays the total cost with a descriptive message.

**5.How much unique type available in shoes**

**6.Which Material shoes in Costly?Justify**

```python
Material = df.groupby('Material')['Price'].mean().idxmax()
print('Material with the highest average price:',Material)
```

This code identifies and prints the material with the highest average price in the DataFrame df. It first
groups the DataFrame by the 'Material' column and calculates the mean price for each material using
.mean(). The .idxmax() function then finds the material with the highest average price. The print()
statement displays this material with a descriptive message, highlighting which material tends to have
the highest average price.

**7.How many days are remaining to released the shoes from current date**

```python
from datetime import datetime
df['Release_Date'] = pd.to_datetime(df['Release_Date'],format='%d-%m-%Y')
fixed_date = datetime(2023,1,1)
df['Days_Remaining'] = (df['Release_Date'] - fixed_date).dt.days
df.head(10)
```

This code converts the 'Release_Date' column in the DataFrame df to datetime format and calculates
the number of days from January 1, 2023, to each release date. It creates a new column,
'Days_Remaining', to store these values, and then displays the first 10 rows of the updated DataFrame.

'Price' in descending order and selects the top entry using .head(1). The print() statement displays the
details of this most expensive shoe.

**11.Perform the cross-tabulation an gender & material & explain**

```python
cross_tab = pd.crosstab(df['Gender'], df['Material'])
print(cross_tab)
```

This code creates and prints a cross-tabulation table that shows the frequency distribution of shoes by
'Gender' and 'Material' in the DataFrame df. The pd.crosstab() function calculates how many shoes of
each material are associated with each gender, resulting in a matrix where the rows represent genders
and the columns represent materials. The print() statement displays this table, providing insight into
the distribution of different materials across genders.
